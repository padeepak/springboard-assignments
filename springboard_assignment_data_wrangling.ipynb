```python
# After checking the dataframe carefully it looks like the column MIN_TEMP_GROUND has acceptable value after every 5 
#occurence i.e. the 6 row in the column  . Below operation create a pandas series to generate a list of numbers 
#with values incrementing by 6 . 

#We would pass the Series to dataframe to get the row where MIN_TEMP_GROUND has no Null. 
sixth_row = pd.Series(range(5,len(df2['MIN_TEMP_GROUND']),6))
df2.loc[sixth_row].head()
```

```python
# The other way to handle the Null values in MIN_TEMP_GROUND is to use FILL method . But first we will drop 
#some unnecessary columns whihc has Null values but has no use for us in general .  
# usage for Drop mentioned below which would delete the entire column and inplace would make 
#it permanent for your dataset
####df2.drop('WEATHER_CODE',axis=1,inplace=True)
```

```python
df2.head()
```

```python
# now to fill the values for MIN_TEMP column to handle the Null value . 
# frontfill operation would take up the values at 6th row and use the same to fill the column values after it . In our
# case though it would keep the first 5 values Null . 
df2['MIN_TEMP_GROUND'].fillna(method='ffill').head(12)
```

```python
# backfill operation would take up the values at 6th row and use the same to fill the column values before it . 
df2['MIN_TEMP_GROUND'].fillna(method='bfill',inplace=True) 
```

```python
# Now we need to concentrate on the other columns where we still see the True for Null such as CLOUD,MIST etc. 
df2.isnull().any()
```

```python
# the values shows Nan for most of the columns in the end ..i.e. cloud ,mist . There are two ways to handle the same .
#either drop the same or use fillna method with the appropriate value for the columns . 
df2[df2.isnull().any(axis=1)].head()
```

```python
# to drop the columns ,it is pretty straightforward : either we can drop all the NAN values in the dataframe 
#or we can target the specific columns .
# it will delete the entire NAN . 
# created a copy here of df2 just to not touch the original dataframe . 
df3 = df2.copy()
#df2.dropna(inplace=True) 
df3.dropna(subset = ['VIEW_RANGE','CLOUD','MIST','RAIN','SNOW','THUNDER','ICE'],inplace = True)
df3.isnull().any() # and no null value is presend in the dataframe . 
```
