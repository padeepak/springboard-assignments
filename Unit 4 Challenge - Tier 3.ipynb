#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_name = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_name, sheet_name='Average price', index_col= None)
properties.head()
```

```python
# fetch the months from the YEar column 
properties['Years']= properties['Year'].dt.year
```

```python
#Applying pivot table to aggreage the columns on the basis of year to understand the price change per year . 
group_pivot = properties.pivot_table(values = name ,index= 'Years',aggfunc = sum )/10000
```

```python
# tag a new column year to use the same while plotting . 
group_pivot['Years'] = group_pivot.index
group_pivot.info()
```

```python
new_df =group_pivot[ (group_pivot['Years'] ==2020) | (group_pivot['Years'] == 1995 )] 
new_df = new_df.drop('Years',axis=1 )
```

Idea here is to get the columns of 1995 and 2005 price values and subtract 1995 values from 2005 to get the change in price for the decade  

```python
def difference(element) :
    return (element.iloc[1] - element.iloc[0])

price_revision = difference(new_df[name])
new_df = new_df.append(price_revision,ignore_index = True)    

new_df
    
```
